# 生物医学信号处理 实验报告

```python
# -*- coding: utf-8 -*-
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

取样频率为1500Hz，选取10s长度的信号

```python
fr = open("data.txt")

dataArray = []
for line in fr.readlines():
    dataArray.append(float(line))
    
data = np.array(dataArray)
x = np.linspace(0,10,data.size)
print(data.size)
```

```python
plt.figure(figsize=(16,8))
plt.plot(x, data)
plt.title(u'选取信号的波形')
```

包络线

```python
import scipy.signal as signal
from scipy import fftpack
```

```python
x = x[:1500]
data = data[:1500]
data_H = fftpack.hilbert(data)
plt.figure(figsize=(16,8))
plt.plot(data)
plt.plot(np.sqrt(data**2 + data_H**2), 'r', linewidth=0.5)
```

频谱图

```python
sampling_rate = 1500
fft_size = 512

dataCutted = data[:fft_size]
data_F = np.fft.rfft(dataCutted)/fft_size
freqs = np.linspace(0, sampling_rate/2, fft_size/2+1)
data_FP = 20*np.log10(np.clip(np.abs(data_F), 1e-20, 1e100))
plt.figure(figsize=(16,8))
plt.subplot(211)
plt.plot(x, data)
pl.xlabel(u"时间(秒)")
pl.title(u"156.25Hz和234.375Hz的波形和频谱")
pl.subplot(212)
pl.plot(freqs, data_FP)
pl.xlabel(u"频率(Hz)")
pl.subplots_adjust(hspace=0.4)
```
